## Creating test datasets using sklearn.datasets.sample_generator

## Blobs Making

```
from sklearn.datasets.samples_generator import make_blobs
#Importing matplotlib for ploting
from matplotlib import pyplot as plt
from matplotlib import style
```

```
#all the stylesheet available in pyton
print(plt.style.available)
```

```
X,y = make_blobs(n_samples=100 , n_features=2 , centers=3 ,cluster_std=1)
```

```
plt.scatter(X[:,1],X[:,0],s=10,color='r')
plt.xlabel("X")
plt.ylabel("Y")
```

## Moon making

```
from sklearn.datasets.samples_generator import make_moons
from matplotlib import pyplot as plt
from matplotlib import style
```

```
X,Y=make_moons(n_samples=1000, noise=0.1)
#change the noise for fun
```

```
plt.scatter(X[:,0],X[:,1],s=10,color='r')
plt.xlabel("X")
plt.ylabel("Y")
```
